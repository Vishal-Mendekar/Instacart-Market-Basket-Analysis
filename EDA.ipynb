# Instacart Market Basket Analysis

Data Source: https://www.kaggle.com/c/instacart-market-basket-analysis/data <br>

Instacart is one of the major grocery ordering and delivery application in US and Canada with headquarters based in San Francisco.

Instacart's aim is to make grocery ordering and delivery easier for its customer by making it available with in just touch of buttons. As the customer places the order the order is reviewed by the personal shoppers and they do indoor shopping and then is delivered to the customer.

Instacart introduces there public dataset back in 2017 in Kaggle platform. Challenging the kaggler to solve the problem as a Classification Problem by telling about the reordering of a specific product by a specific user.

The publically available dataset consist of the data sample of over 3 million grocery orders from more than 200,000 Instacart users.

### Attributes Of Each CSV files are given below 

<h3>1. Aisles.csv</h3>

1. Aisle Id : A unique Id to represent each aisle (Integer : int16)

2. Aisle : Contains the name of aisle based on products on aisle (String)

<h3>2. Departments.csv</h3>

1. Department Id : Unique integer to represent each department (Integer : int16)

2. Department : String which tells the name of department depending upon products in department (String)

<h3>3. Order Products Prior.csv</h3>

1. Order Id : Unique integer to represent order (Integer : int)

2. Product Id : Unique integer for each product (Integer : int32)

3. Add to cart order : Order in which product is added in the cart (Integer : int16)

4. Reordered : Binary variable (0 = Not reordered /1 = Reordered)
               
<h3>4. Order Products Train.csv</h3>   

1. Order Id : Unique integer to represent order (Integer : int)

2. Product Id : Unique integer for each product (Integer : int32)

3. Add to cart order : Order in which product is added in the cart (Integer : int16)

4. Reordered : Binary variable (0 = Not reordered /1 = Reordered)

<h3>5. Order.csv</h3>

1. Order Id : Unique integer to represent order (Integer : int64)

2. User Id : Unique integer to represent different users (Integer : int64)

3. Eval Set : Tells whether order is from Prior or Train (String)

4. Order Number : Order number for the order made by customer (Integer : int64)

5. Order Dow : Ranges from 0-6 where 0 =  Sunday and 6 = Saturday (Integer : int16)

6. Order Hour Of Day : Ranges from 0-23 where 0 = 12 a.m and 23 = 11 p.m (Integer : int16)

7. Day Since Prior Order : Number of days since last order is placed (Integer : int16)

<h3>6. Products.csv</h3>

1. Product Id : Unique Id for each product since there are huge number of products so will have large range (Integer : int64)

2. Product Name : Name of the product (String)

3. Aisle Id : Id of the aisle where the product is present (Integer : int64)

4. Department Id : Unique Id for each department (Integer : int32)

<h3>7. Sample Submission.csv</h3>

1. Order Id : Id for each order placed (Integer : int64) 

2. Products : List of space delimited product id which all products will be ordered (Integer : int64)


### Objective:

Given user id we will have to tell whether that certain product will be ordered by the user or not. And as results in test set will have to return list of all such product id's.


### Performance Matrices : F1 Score (given)

<b>[Q]</b> How to predict whether item will be ordered by user or not?

<b>[A]</b> Given history of each user we can observe some patterns like on which day and on what time certain user order some products. Or what other products are added to the cart after adding a specific product.

One such pattern seen in the orders is that on thrusday night beer and baby pads are ordered frequently. Maybe because dad's want to stock up for weekend and enjoy during weekends.
